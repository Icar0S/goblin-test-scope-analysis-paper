```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file and adjust columns
df_cat = pd.read_csv(r"data\Artefatos com mais dependencia (Por biblioteca).csv", skiprows=1)
df_cat.columns = ["Artifact", "Library", "ID", "Versions", "CVEs_Presence", "Total_CVEs", 
                  "Reference", "Ranking", "Category", "Used_by"]

# Convert numeric types
df_cat["Total_CVEs"] = pd.to_numeric(df_cat["Total_CVEs"], errors='coerce').fillna(0).astype(int)
df_cat["Used_by"] = pd.to_numeric(df_cat["Used_by"], errors='coerce').fillna(0).astype(int)

# Check that data is present
print(df_cat[["Library", "Total_CVEs", "Used_by"]])

# Plot scatter: Total_CVEs vs Used_by
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cat, x="Used_by", y="Total_CVEs", hue="Category", s=80)
plt.title("CVEs vs Popularity (Used by Other Artifacts)", fontsize=16, fontweight='bold')
plt.xlabel("Number of Artifacts Using", fontsize=14)
plt.ylabel("Total Number of CVEs", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="Category", fontsize=12, title_fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset (ajuste o caminho, se necessrio)
df = pd.read_csv(r"data\3columnsCVEs-t30.csv")

# Exibe os tipos de dados originais
print("Cabealho:")
print(df.head())

# Criar uma nova coluna numrica: 0 para "No CVE", 1 para qualquer outra entrada
df["CVE_indicator"] = df["CVEs"].apply(lambda x: 0 if x.strip().lower() == "no cve" else 1)

# Converter a coluna Timestamp para numrico, se necessrio
df["Timestamp"] = pd.to_numeric(df["Timestamp"], errors="coerce")

# Selecionar apenas as colunas numricas relevantes
df_numeric = df[["Timestamp", "CVE_indicator"]]

print("Colunas numricas disponveis:", df_numeric.columns)
print("Forma do dataframe numrico:", df_numeric.shape)

# Calcular as matrizes de covarincia e correlao
cov_matrix = df_numeric.cov()
corr_matrix = df_numeric.corr()

# Plotar o heatmap da matriz de covarincia
plt.figure(figsize=(6, 5))
sns.heatmap(cov_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Covarincia")
plt.tight_layout()
plt.show()

# Plotar o heatmap da matriz de correlao
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlao")
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar o dataset
df = pd.read_csv(r"data\3columnsCVEs-t30.csv")

# 2. Criar uma funo para extrair a parte numrica inicial da verso
def parse_version(version_str):
    """
    Ex.: "4.13-beta-1" -> 4.13
         "3.8" -> 3.8
         "4.1" -> 4.1
         Qualquer coisa que no casar -> None
    """
    match = re.match(r'^(\d+(\.\d+)?)', str(version_str))
    if match:
        return float(match.group(1))
    return None

# 3. Criar coluna numrica para a verso
df["Version_num"] = df["Version"].apply(parse_version)

# 4. Criar indicador 0/1 para CVE
df["CVE_indicator"] = df["CVEs"].apply(lambda x: 0 if x.strip().lower() == "no cve" else 1)

# 5. Converter o Timestamp (ms) em "ano aproximado"
#    Dividimos por 1000 -> segundos,
#                /3600 -> horas,
#                /24   -> dias,
#                /365.25 -> anos,
#    depois somamos 1970 para alinhar ao calendrio
df["Year"] = df["Timestamp"] / 1000 / 3600 / 24 / 365.25 + 1970

# 6. Filtrar linhas que tenham valor de verso parseada
df_filtered = df.dropna(subset=["Version_num"])

# 7. Plotar scatter: Eixo X = "Year", Eixo Y = "Version_num", cor = "CVE_indicator"
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_filtered, x="Year", y="Version_num", hue="CVE_indicator", palette="Set1", s=50)
plt.title("Distribuio de Verses ao Longo do Tempo (CVE x No-CVE)")
plt.xlabel("Ano (estimado a partir de Timestamp)")
plt.ylabel("Verso parseada (ex.: 4.13 -> 4.13)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"data\3columnsCVEs-t30.csv")

# Display the header for verification
print("Header:")
print(df.head())

# Convert Timestamp (milliseconds since 1970) to Year.
# Divide by: 1000 (ms to s), 3600 (s to hours), 24 (hours to days), and 365.25 (days to years), then add 1970.
df["Year"] = df["Timestamp"] / 1000 / 3600 / 24 / 365.25 + 1970
df["Year"] = df["Year"].astype(int)

# Create a new column "CVE_count": assign 0 if 'CVEs' equals "No CVE", else 1.
df["CVE_count"] = df["CVEs"].apply(lambda x: 0 if str(x).strip().lower() == "no cve" else 1)

# Aggregate the data by Year to count the total number of CVEs each year
cve_by_year = df.groupby("Year")["CVE_count"].sum().reset_index()

# Display the aggregated data for confirmation
print("Aggregated CVEs by Year:")
print(cve_by_year)

# Plot the area/line chart
plt.figure(figsize=(10, 6))
plt.plot(cve_by_year["Year"], cve_by_year["CVE_count"], marker='o', color='blue', linewidth=2)
plt.fill_between(cve_by_year["Year"], cve_by_year["CVE_count"], color='skyblue', alpha=0.4)
plt.title("Number of CVEs Over Time", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Total Number of CVEs", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
